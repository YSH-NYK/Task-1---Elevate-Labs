# Removing null, na and duplicate data

```python
df = df.drop_duplicates() #Not required for my dataset as it has no duplicates, but added for future refrence
```

```python
df['director'] = df['director'].fillna("Unknown")
df['cast'] = df['cast'].fillna("Unknown")
df['country'] = df['country'].fillna("Unknown")
df['rating'] = df['rating'].fillna("Not Rated")
```

```python
#converting all dates from August 14, 2020 format to 14-08-2020 format
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce') 
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')
# Adding date values as 01-01-2020 for na and null data
df['date_added'] = df['date_added'].fillna("01-01-2020")
```

# Renamed column headers to be clean and uniform

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_')
```

```python
df.columns
```

# Remove leading/trailing spaces

```python
df['type'] = df['type'].str.strip().str.title()       # "movie" → "Movie"
df['country'] = df['country'].str.strip().str.title() # "united states " → "United States"
df['rating'] = df['rating'].str.strip().str.upper()   # "tv-ma " → "TV-MA"
```

```python
df.head()
```

# Replace inconsistent Country Names

```python
country_map = {
    "United States Of America": "United States",
    "Usa": "United States",
    "Uk": "United Kingdom",
}
df['country'] = df['country'].replace(country_map)
```

```python
df.head()
```

# Standardized Ratings 

```python
rating_map = {
    "Tv-Ma": "TV-MA",
    "Tv14": "TV-14",
    "Pg ": "PG",
    "Nr": "Not Rated"
}

df['rating'] = df['rating'].replace(rating_map)
```
